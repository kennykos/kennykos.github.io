```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
# Define the sinc function
x = np.linspace(-4 * np.pi, 4 * np.pi, 1000)
sinc = np.sinc(x / np.pi)  # numpy's sinc is normalized as sin(pi*x)/(pi*x)

# Compute the Fourier Transform
freq = np.fft.fftfreq(len(x), d=(x[1] - x[0])) * 2 * np.pi  # Angular frequency
sinc_ft = np.fft.fft(sinc)
sinc_ft = np.fft.fftshift(sinc_ft)  # Shift zero frequency to the center
freq = np.fft.fftshift(freq)  # Align frequency axis

# Normalize the Fourier Transform for better scaling
sinc_ft_magnitude = np.abs(sinc_ft)
sinc_ft_magnitude /= sinc_ft_magnitude.max()

# Plot the sinc function and its Fourier Transform
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot sinc function
axs[0].plot(x, sinc, label='sinc(x)', color='blue')
axs[0].set_title("Sinc Function")
axs[0].set_xlabel("time")
axs[0].set_ylabel("Amplitude")
axs[0].set_xticks(np.arange(-4 * np.pi, 5 * np.pi, np.pi))
axs[0].set_xticklabels(["$-4\\pi$", "$-3\\pi$", "$-2\\pi$", "$\\pi$", "$0$", "$\\pi$", "$2\\pi$", "$3\\pi$", "$4\\pi$"])
axs[0].grid(True)
axs[0].legend()

# Plot Fourier Transform
axs[1].plot(freq, sinc_ft_magnitude, label='$|\\mathcal{F}(\\mathrm{sinc})(x)|$', color='red')
axs[1].set_title("Fourier Transform of Sinc Function")
axs[1].set_xlabel("Frequency")
axs[1].set_ylabel("Magnitude")
axs[1].grid(True)
axs[1].legend()

# Adjust layout and show
plt.tight_layout()
plt.show()
```
